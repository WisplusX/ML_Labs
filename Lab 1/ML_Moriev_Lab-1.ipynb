```python
import numpy as np
import matplotlib.pyplot as plt

# Читання даних з файлу .csv
train = np.genfromtxt('Lab1_data/lab_1_train.csv', delimiter=',', skip_header=1)
test = np.genfromtxt('Lab1_data/lab_1_test.csv', delimiter=',', skip_header=1)

# Розділення даних на стовпці x та y
x_train = train[:, 1]
y_train = train[:, 2]

x_test = test[:, 1]
y_test = test[:, 2]
```

```python
# Дані для тренування моделі
x_train, y_train
```

```python
# Дані для перевірки моделі
x_test, y_test
```

```python
# Візуалізація навчальних даних
plt.plot(x_train, y_train, 'o', markersize=3, color='green')  # 'o' - рисувати точки, markersize - розмір маркера
plt.xlabel('x')  # підпис осі x
plt.ylabel('y')  # підпис осі y
plt.title('Train Data')  # заголовок графіка
plt.grid(True)  # ввімкнення сітки
plt.show()  # показати графік
```

```python
# Візуалізація тестових даних
plt.plot(x_test, y_test, 'o', markersize=3, color='purple')  # 'o' - рисувати точки, markersize - розмір маркера
plt.xlabel('x')  # підпис осі x
plt.ylabel('y')  # підпис осі y
plt.title('Test Data')  # заголовок графіка
plt.grid(True)  # ввімкнення сітки
plt.show()  # показати графік
```

```python
# Прогнозуємо значення за допомогою лінійної регресії
def calculate_linear_regression(intercept, slope, x):
    return intercept + slope * x

# Обчислюємо середньоквадратичну помилку між фактичними та прогнозованими значеннями
def calculate_mean_squared_error(y_true, y_pred):
    return np.mean((np.array(y_pred) - np.array(y_true)) ** 2)

# Маємо таку функцію для розрахунку градієнту
def calculate_mean_squared_error_derivatives(y_true, y_pred, x):
    difference_y = y_true - y_pred
    der_intercept = - (2 / len(difference_y)) * np.sum(difference_y)
    der_slope = - (2 / len(difference_y)) * np.sum(np.multiply(difference_y, x))
    return der_intercept, der_slope
```

```python
# Розробимо функцію для навчання моделі
def train_linear_regression_model(x_train, y_train):
    # Точність, з якою ми хочемо досягти збіжності
    tolerance = 0.000001
    
    # Швидкість навчання (learning rate)
    lr = 0.01

    # Ініціалізація випадковими значеннями для перехоплення та нахилу
    intercept, slope = np.random.normal(), np.random.normal()

    # Обчислення прогнозованих значень та початкової функції втрат
    y_pred = calculate_linear_regression(intercept, slope, x_train)
    loss = calculate_mean_squared_error(y_train, y_pred)
    
    # Початкові значення для порівняння різниці втрат та лічильника епох
    dt = 1000
    i = 0

    # Градієнтний спуск
    while dt > tolerance:
        # Обчислення похідних від функції втрат
        der_intercept, der_slope = calculate_mean_squared_error_derivatives(y_train, y_pred, x_train)

        # Оновлення значень перехоплення та нахилу
        intercept -= lr * der_intercept
        slope -= lr * der_slope

        # Переобчислення прогнозованих значень та функції втрат
        y_pred = calculate_linear_regression(intercept, slope, x_train)
        cur_loss = calculate_mean_squared_error(y_train, y_pred)

        # Обчислення різниці втрат
        dt = np.abs(loss - cur_loss)
        loss = cur_loss
        i += 1

        # Виведення інформації про кожну епоху
        print(f"{i}: Loss = {loss}. Intercept = {intercept}. Slope = {slope}")

    return {"intercept": intercept, "slope": slope}
```

```python
# Викличемо її і подивимось на результат
result = train_linear_regression_model(x_train, y_train)
```

```python
# Обчислюємо середньоквадратичну помилку для тестового набору даних,
# використовуючи прогнозовані значення лінійної регресії з вагами,
# які були отримані після навчання моделі.
loss = calculate_mean_squared_error(y_test, calculate_linear_regression(result["intercept"], result["slope"], x_test))
print("Loss = ", loss)
```

```python
# Візуалізація тренувальних і тестових даних та лінії регресії
plt.scatter(x_train, y_train, color='green', label='Train Data')
plt.scatter(x_test, y_test, color='purple', label='Test Data')
plt.plot( [0, 1], [calculate_linear_regression(result["intercept"], result["slope"], 0), calculate_linear_regression(result["intercept"], result["slope"], 1)], color='black', linewidth=3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
